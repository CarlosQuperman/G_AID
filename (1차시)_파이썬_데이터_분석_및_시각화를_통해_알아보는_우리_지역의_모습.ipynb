### 인구 분석 및 인구 변화 시각화

* 데이터 수집처 : https://jumin.mois.go.kr/


```python
#강원 지역 시군구별 인구 변화 데이터 불러오기

import pandas as pd
import plotly.express as px

kang_pop = pd.read_csv("/content/강원도 시군구별 인구 변화(편집)_2008-2024.csv",encoding='cp949')
display(kang_pop.head())
display(kang_pop.info())
display(kang_pop['행정구역'].unique())
display(kang_pop.describe())
```

```python
#강원 지역 시군구별 인구 변화 애니메이션 시각화

kang_pop_ani = px.bar(kang_pop,x="행정구역",y="총인구수",animation_frame="연도",animation_group="행정구역",color="행정구역")
kang_pop_ani
```
